```python
#Importing needed modules .
from pandas import read_csv , DataFrame , concat
from numpy import nan , array
from sklearn.preprocessing import StandardScaler 
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
#Instancing from 'StandardScaler' and 'SimpleImputer' objects .
scaler = StandardScaler()
imputer = SimpleImputer()
```

```python
#Reading csv file and getting 10 sample of that randomly .
file_path = "D:\\datasets\\water quality\\waterQuality1.csv"
file = read_csv(file_path)
df = file.copy()
df.sample(10)
```

```python
#Getting information from data frame .
df.info()
```

```python
#Finding '#NUM!' value in target column and drop them from data frame .
temp = list()
for i in df.to_numpy() :
    if i[-1] == '#NUM!' :
        continue
    temp.append(i)
df = DataFrame(df,columns=df.columns)
```

```python
#There are some '#NUM!' values in 'ammonia' column , so lets replace them by mean strategy .
ammonia = df.ammonia
#Seperating labels .
lbl = df.is_safe
df = df.drop(['ammonia','is_safe'],axis=1)
```

```python
new_ammonia = list()
for i in ammonia :
    if i == '#NUM!' :
        new_ammonia.append(nan)
    else :
        new_ammonia.append(float(i))

reshaped_ammonia = array(new_ammonia).reshape(-1,1)
new_ammonia = DataFrame(imputer.fit_transform(reshaped_ammonia),columns=['Ammonia'])
```

```python
#Attaching imputed 'ammonia' to main datas .
df = concat([df,new_ammonia],axis=1)
df.isnull().sum()
```

```python
#Scaling datas .
scaled_df = DataFrame(scaler.fit_transform(df),columns=df.columns)
scaled_df.describe()
```

```python
#Seperating datas to train and test sets .
xtr , xts , ytr , yts = train_test_split(scaled_df,lbl,test_size=0.2,random_state=42)
```

```python
message = 'There are %d samples in train set and %d samples in test set .'%(len(xtr),len(xts))
print(message)
```

```python
#Instancing from 'DecisionTreeClassifier' object and fit it with train data set .
model = DecisionTreeClassifier().fit(xtr,ytr)
```

```python
#Calculating scores .
model.score(xtr,ytr)*100
```
